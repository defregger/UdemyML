```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4])
y = np.array([8,15,22,31])

```

```python
plt.plot(x, y, color="red")
plt.xlabel("x-achse")
plt.ylabel("y-achse")
plt.show()
```

```python
x = np.arange(start=-20, stop=21)
y = x**2

print(x)
print(y)

plt.scatter(x, y, c="purple")
plt.show()
```
