```python
# langer Weg
my_list = []

for i in range(10):
    my_list.append(i)
print(my_list)
```

```python
# besser und schneller mit comprehensiom

my_list2 = [i for i in range(10)]
print(my_list2, "\n")

my_list3 = [i**2 if i % 2 == 0 else -1 for i in range(20)]
print(my_list2, "\n")

```

```python
# Slicing
#                     START:STOP-1  
my_list_sl = my_list3[:3] #startet bei index 0 bis 2 
print(my_list_sl, "\n")

my_list_sl = my_list3[4:8] #startet bei index 0 bis 2 
print(my_list_sl, "\n")
```

```python
import numpy as np 
import matplotlib.pyplot as plt


my_array = np.arange(16).reshape(8,2)
print("my-array: ",my_array, "\n")

x_co = my_array[:,0]
y_co = my_array[:,1]
print("X-Koordinate",x_co, "\n")
print("Y-Koordinate", y_co, "\n")

plt.scatter(x_co,y_co)
plt.show()
```

```python

print("my-array: ",my_array, "\n")

x_co[0] = -10 # vVerändert die Matrix!!!


print("my-array: ",my_array, "\n")

plt.scatter(x_co,y_co)
plt.show()
```
