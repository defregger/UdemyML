```python
# Aufgabe 1
# zuf√§lligen, ganzzahligen Datensatz als Numpy Array mit der Funktion np.random.randint() anlegen
# 
import numpy as np
import matplotlib.pyplot as plt

size = 30

my_arr = np.random.randint(0, 100, size=(size), dtype=int)
print(my_arr, "\n")
```

```python
# Aufgabe 2
# stelle my_arr in Matplotlib als Scatterplot dar, bei dem Y-Koordinaten die Werte des Array sind und die X-Koordinaten die  Indizies der Listen-Werte sind 
# 
plt.scatter(range(size), my_arr)

# Aufgabe 3 Achsenbeschriften
plt.ylabel("Y-Axis")
plt.xlabel("X-Axis")


# Aufgabe 4 Mittelwert
my_arr_av = np.full(shape=(size), fill_value=np.mean(my_arr))
plt.plot(range(size), my_arr_av, color="red")
print(my_arr_av)

# Aufgabe 5 Median
my_arr_med = np.full(shape=(size), fill_value=np.median(my_arr))
plt.plot(range(size), my_arr_med, color="yellow")

plt.show()
```
