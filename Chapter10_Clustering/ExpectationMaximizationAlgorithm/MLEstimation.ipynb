#### Normal Distribution

$f(x | \mu, \sigma^2) = \frac{1}{\sqrt{2 \pi \sigma^2}} \exp{\bigg(-\frac{(x-\mu)^2}{2\sigma^2}\bigg)}$$f(x | \mu, \sigma^2) = \frac{1}{\sqrt{2 \pi \sigma^2}} \exp{\bigg(-\frac{(x-\mu)^2}{2\sigma^2}\bigg)}$

#### Maximum Lilihood Estimation

$ln(\mathcal{L}(\mu, \sigma^2)) = 
\sum_{i = 1}^n ln(f(x_i \ | \ \mu, \sigma^2))$  

$ = ln\bigg(\frac{1}{\sqrt{2 \pi \sigma^2}}\bigg) + ln\bigg(exp\bigg(- \frac{1}{2\sigma^2} \sum_{i = 1}^n (x_i-\mu)^2\bigg)\bigg)$  

$= -\frac{n}{2} ln\big(2 \pi\big)  -\frac{n}{2} ln\big(\sigma^2\big) - \frac{1}{2\sigma^2} \sum_{i = 1}^n (x_i - \mu)^2$  

#### MLE Computation

$\hat{\mu} = \frac{1}{n} \sum_{i = 1}^n x_i$  
  
$\hat{\sigma}^2 = \frac{1}{n} \sum_{i = 1}^n (x_i - \hat{\mu})^2$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def generate_dataset():
    std = 3
    mean = 1.5
    data = np.random.normal(mean, std, 10)
    return data
```

```python
data = generate_dataset()

plt.scatter(data, [0 for i in range(len(data))])
plt.show()
```

$\hat{\mu} = \frac{1}{n} \sum_{i = 1}^n x_i$  
  
$\hat{\sigma}^2 = \frac{1}{n} \sum_{i = 1}^n (x_i - \hat{\mu})^2$

```python
def estimate_mean(data):
    mean = np.mean(data)
    return mean
```

```python
def estimate_std(data):
    std = np.std(data)
    return std
```

```python
mean = estimate_mean(data)
std = estimate_std(data)

print("Mean: ", mean)
print("Std: ", std)
```

```python
real_pdf = [norm.pdf(x, 1.5, 3) for x in np.arange(-9, 11, 0.01)]
estimate_pdf = [norm.pdf(x, mean, std) for x in np.arange(-9, 11, 0.01)]
```

```python
plt.scatter(np.arange(-9, 11, 0.01), real_pdf, color="red", s=4)
plt.scatter(np.arange(-9, 11, 0.01), estimate_pdf, color="blue", s=4)
plt.legend(["Y", "$\hat{Y}"])
plt.show()
```

```python
for num_samples in range(100, 11000, 1000):
    data = np.random.normal(1.5, 3, num_samples)
    mean = estimate_mean(data)
    std = estimate_std(data)
    real_pdf = [norm.pdf(x, 1.5, 3) for x in np.arange(-9, 11, 0.01)]
    estimate_pdf = [norm.pdf(x, mean, std) for x in np.arange(-9, 11, 0.01)]
    plt.scatter(np.arange(-9, 11, 0.01), real_pdf, color="red", s=4)
    plt.scatter(np.arange(-9, 11, 0.01), estimate_pdf, color="blue", s=4)
    plt.legend(["Y", "$\hat{Y}$"])
    plt.title("Num Samples = " + str(num_samples))
    plt.show()
```
