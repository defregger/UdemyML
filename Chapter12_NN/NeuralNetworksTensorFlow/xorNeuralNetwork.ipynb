```python
import numpy as np

from tensorflow.keras.models import *
from tensorflow.keras.layers import *
from tensorflow.keras.optimizers import *
from tensorflow.keras.losses import *
from tensorflow.keras.utils import *
```

```python
# XOR Dataset: 
# A, B sind binÃ¤re Werte {0, 1}
# XOR(A, B) = A ^ B
def XOR(A, B):
    if (A == 1 and B == 0) or (A == 0 and B == 1):
        return 1
    else:
        return 0
```

```python
x = np.array([[0,0], [0,1], [1,0], [1,1]], dtype=np.float32)
y = np.array([[0], [1], [1], [0]], dtype=np.float32)
```

```python
model = Sequential()
# Input to first Hidden Layer
model.add(Dense(8, input_dim=x.shape[1])) # Input Layer -> 1. Hidden Layer: 2x8 Weights and 8 biases
model.add(Activation("tanh")) # (-1, +1)
# Hidden to Hidden Layer
model.add(Dense(4))
model.add(Activation("tanh")) # (-1, +1)
# Output Layer
model.add(Dense(y.shape[1]))
model.add(Activation("sigmoid")) # (0, 1): value > 0.5 => class 1, else class 0
model.summary()
```

```python
sgd = SGD(lr=0.1)
model.compile(loss="binary_crossentropy", optimizer=sgd, metrics=["accuracy"])
model.fit(x, y, epochs=200)
```

```python
y_pred = (model.predict(x) > 0.5)
print(y_pred)
```
